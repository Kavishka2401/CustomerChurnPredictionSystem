```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Metrics and values
metrics = ['Accuracy', 'Precision', 'Recall', 'F1 Score', 'ROC-AUC']
nn_values = [0.7559, 0.5284, 0.7460, 0.6186, 0.8369]
dt_values = [0.7480, 0.5179, 0.7353, 0.6077, 0.8167]

x = np.arange(len(metrics))
width = 0.35

nn_color = '#8ecae6'
dt_color = '#90be6d'

# Figure
plt.figure(figsize=(10,6))
bars1 = plt.bar(x - width/2, nn_values, width, label='Neural Network', color=nn_color)
bars2 = plt.bar(x + width/2, dt_values, width, label='Decision Tree', color=dt_color)

# Add value labels on top
for bar, val in zip(bars1, nn_values):
    plt.text(bar.get_x() + bar.get_width()/2, val + 0.01, f'{val:.2f}', ha='center', va='bottom', fontsize=10)
for bar, val in zip(bars2, dt_values):
    plt.text(bar.get_x() + bar.get_width()/2, val + 0.01, f'{val:.2f}', ha='center', va='bottom', fontsize=10)

# Styling
plt.xticks(x, metrics, fontsize=12)
plt.ylim(0, 1.0)
plt.ylabel('Score', fontsize=12)
plt.title('Test Metrics Comparison: Neural Network vs Decision Tree', fontsize=14, fontweight='bold')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```

```python
# Confusion matrices
cm_nn = np.array([[786, 249],
                  [95, 279]])

cm_dt = np.array([[779, 256],
                  [99, 275]])

labels = ['Non-Churn', 'Churn']

# Figure
fig, axes = plt.subplots(1, 2, figsize=(12,5))

# Neural Network
sns.heatmap(cm_nn, annot=True, fmt='d', cmap='Blues', xticklabels=labels, yticklabels=labels, ax=axes[0])
axes[0].set_title('Neural Network')
axes[0].set_xlabel('Predicted')
axes[0].set_ylabel('Actual')

# Decision Tree
sns.heatmap(cm_dt, annot=True, fmt='d', cmap='Oranges', xticklabels=labels, yticklabels=labels, ax=axes[1])
axes[1].set_title('Decision Tree')
axes[1].set_xlabel('Predicted')
axes[1].set_ylabel('Actual')

plt.suptitle('Confusion Matrices Comparison', fontsize=14, fontweight='bold')
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```
