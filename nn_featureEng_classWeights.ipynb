```python
# Import libraries
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Provide path
data=pd.read_csv('/content/drive/MyDrive/processed_data_NN.csv')
```

```python
data_model_2 = data.copy()
```

```python
# 1. Average monthly charge per tenure
# Adding 1 to tenure to avoid division by zero
data_model_2['AvgChargePerMonth'] = data_model_2['TotalCharges_numeric'] / (data_model_2['tenure'] + 1)

# 2. Service usage count
# List of columns representing optional services (already one-hot encoded as 0/1)
service_cols = [
    'OnlineSecurity_Yes', 'OnlineBackup_Yes', 'DeviceProtection_Yes',
    'TechSupport_Yes', 'StreamingTV_Yes', 'StreamingMovies_Yes'
]

# Sum the services used by each customer
data_model_2['ServiceCount'] = data_model_2[service_cols].sum(axis=1)

# Check first few rows to confirm
data_model_2[['AvgChargePerMonth', 'ServiceCount']].head()
```
